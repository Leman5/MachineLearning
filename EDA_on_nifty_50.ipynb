200 day MA on whole period

```python
sns.set_style('whitegrid')
plt.figure(figsize=(18,7))
hdfc_data['High'].rolling(window=200).mean().plot(label='30 day MA on high prices')
hdfc_data['High'].plot(label='high price')
plt.legend()
```

```python
hdfc_data['MA 30day'] = hdfc_data['Close'].rolling(window=30).mean()
hdfc_data['Std 30day'] = hdfc_data['Close'].rolling(window=30).std()
```

```python
hdfc_data['Upper Band'] = hdfc_data['MA 30day'] + (hdfc_data['Std 30day']*2)
hdfc_data['Lower Band'] = hdfc_data['MA 30day'] - (hdfc_data['Std 30day']*2)
```

```python
hdfc_data[['Close','MA 30day','Std 30day','Upper Band','Lower Band']].plot(figsize=(18,7))
plt.title("Bollinger Band Plot for Overall Period", fontsize=17)
```

```python
hdfc_data[['Close','MA 30day','Std 30day','Upper Band','Lower Band']].loc['2017-03-03':'2019-04-03'].plot(figsize=(18,7))
plt.title(" Bollinger Band Plot for 2017-2019", fontsize=17)
```

# Train and test

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
```

```python
X = hdfc_data[['Open','High','Low','Close','Volume','Turnover']]
y = hdfc_data['Close']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0,shuffle=False)
X_test
```

```python
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
X_test
```

```python
model = RandomForestRegressor(n_estimators=500, random_state=42, max_depth=10)
model.fit(X_train, y_train)
predict = model.predict(X_test)
print(predict)
print(predict.shape)
```

```python
print("Mean Absolute Error:", round(metrics.mean_absolute_error(y_test, predict), 4))
print("Mean Squared Error:", round(metrics.mean_squared_error(y_test, predict), 4))
print("Root Mean Squared Error:", round(np.sqrt(metrics.mean_squared_error(y_test, predict)), 4))
print("(R^2) Score:", round(metrics.r2_score(y_test, predict), 4))
print(f'Train Score : {model.score(X_train, y_train) * 100:.2f}% and Test Score : {model.score(X_test, y_test) * 100:.2f}% using Random Tree Regressor.')
errors = abs(predict - y_test)
mape = 100 * (errors / y_test)
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.') 
```
